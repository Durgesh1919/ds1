```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# LINE PLOT
plt.plot([1,4,8,10])
plt.ylabel("Y-axis Numbers")
plt.xlabel("X-axis Numbers")
plt.show()
```

```python
# BAR PLOT
names = ["WAD","DSBDA","DSA"]
values = [25,45,65]
plt.bar(names,values)
plt.show()
```

```python
# SCATTER PLOT
names = ["WAD","DSBDA","DSA"]
values = [25,45,65]
plt.scatter(names,values)
plt.show()
```
